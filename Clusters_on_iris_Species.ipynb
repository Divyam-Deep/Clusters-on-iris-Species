```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/iris_dataset.csv')
data.head()
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'],data['petal_length'],data['petal_width'])
plt.show()
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster'] = kmeans.fit_predict(x)
clusters
```

```python
plt.scatter(clusters['sepal_length'],clusters['sepal_width'], c = clusters['cluster'], cmap = 'rainbow')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(i)
  kmeans.fit(x_scaled)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans_new = KMeans(6)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['Clusters'] = kmeans_new.fit_predict(x_scaled)
clusters_new.head()
```

```python
plt.scatter(clusters_new['sepal_length'], clusters_new['sepal_width'], c=clusters_new['Clusters'], cmap='rainbow')
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['Clusters'] = kmeans_new.fit_predict(x_scaled)
clusters_new.head()
```

```python
plt.scatter(clusters_new['sepal_length'], clusters_new['sepal_width'], c=clusters_new['Clusters'], cmap='rainbow')
plt.show()
```
